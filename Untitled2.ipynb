```python
# 필요 라이브러리 호출
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# 입력 데이터 불러오기
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/dataset.CSV', encoding='cp949')
```

```python
# 입력 데이터 확인(처음 5행 출력)
data.head()
```

```python
#입력 데이터 정보 확인
data.info()
```

```python
#기준 데이터 준비(국내 화물차량 5종 차량제원)

Width = [1.55, 2.00, 2.20, 2.34, 2.34]
Height = [2.40, 2.90, 3.70, 3.90, 3.90]
Length = [2.80, 4.30, 6.20, 9.10, 10.10]

df = pd.DataFrame({'Width(m)' : Width, 'Height(m)' : Height, 'Length(m)' : Length })
df.index = ['1톤 트럭', '3.5톤 트럭', '5톤 트럭', '11톤 트럭', '25톤 트럭']

#기준 데이터 확인
df
```

```python
# 통행 불가능한 차량제원의 개수를 반환하는 함수 정의
def count_unpassable_roads(width, height, length):
    unpassable_count = 0

    # 각 도로에 대해 통행 불가능 여부 확인(기준 데이터와의 비교)
    for idx, truck in df.iterrows():
        if truck['Width(m)'] >= width or truck['Height(m)'] >= height or truck['Length(m)'] >= length:
            unpassable_count += 1

    return unpassable_count

# 입력 데이터에서 'Width(m)', 'Height(m)', 'Length(m)' 열 추출
selected_data = data[['Width(m)', 'Height(m)', 'Length(m)']].values

# 각 구간별로 통행 불가능한 차량제원의 개수 계산
unpassable_road_counts = []
for row in selected_data:
    unpassable_road_counts.append(count_unpassable_roads(row[0], row[1], row[2]))

# 결과 출력
result_df = pd.DataFrame({'구간 번호': data.index, '통행 불가능한 차량 제원의 개수': unpassable_road_counts})
print(result_df)
```

```python
# 차량제원별 통행 불가능한 도로 개수 계산
truck_types = ['모든 화물차량 통행 가능', '25톤 트럭 이상', '11톤 트럭 이상', '5톤 트럭 이상', '3.5톤 트럭 이상', '1톤 트럭 이상']
cumulative_counts = Counter()
total_count = 0
counts = Counter(unpassable_road_counts)

for num_roads, count in counts.items():
    total_count += count
    cumulative_counts[num_roads] = total_count

# 결과 출력
for num_roads, cumulative_count in cumulative_counts.items():
  if num_roads == 0:
        print(f"{truck_types[0]}한 도로는 {cumulative_count}개 입니다.")
  else:
        print(f"{truck_types[num_roads]}이 통행 불가능한 도로는 {cumulative_count}개 입니다.")
```

```python
# 차량제원별 통행 불가능한 도로 개수 막대 그래프
truck_counts = {truck_types[num_roads]: cumulative_count for num_roads, cumulative_count in cumulative_counts.items()}

# 막대 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(truck_counts.keys(), truck_counts.values(), color='skyblue')
plt.xlabel('트럭 종류')
plt.ylabel('통행 불가능한 도로 개수')
plt.title('차량제원별 통행 불가능한 도로 개수')
plt.tight_layout()
plt.show()
```

```python
# 차량제원별 통행 불가능한 도로 백분율 계산

# 결과 출력
percentages = {}
for num_roads, cumulative_count in sorted(cumulative_counts.items()):
  if num_roads != 0:
    percentage = cumulative_count / total_count * 100
    percentages[num_roads] = percentage
    print(f"{truck_types[num_roads]}이 통행 불가능한 도로는 {cumulative_count}/{total_count}, 백분율은 {percentage:.2f}% 입니다.")
```

```python
# 차량제원별 통행 불가능한 도로 백분율 그래프

percentages_dict = {truck_types[num_roads]: percentage for num_roads, percentage in percentages.items()}

# 막대 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(percentages_dict.keys(), percentages_dict.values(), color='blue')
plt.xlabel('트럭 종류')
plt.ylabel('통행 불가능한 도로 백분율')
plt.title('차량제원별 통행 불가능한 도로 백분율')
plt.tight_layout()
plt.show()
```
